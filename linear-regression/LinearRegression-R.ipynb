# Linear regression demo

```r
library(tidyverse)
options(repr.plot.width=4, repr.plot.height=4)
```

```r
simulated_data <- data.frame('x' = rnorm(100), 'y' = rnorm(100))

ggplot(simulated_data, aes(x=x,y=y)) + geom_point()
```

## $ y = a + b \cdot x $

```r
model <- lm(y~x, simulated_data)
summary(model)  
```

```r
simulated_data2 <- rbind(simulated_data, data.frame(x = 100, y = 100))

ggplot(simulated_data2, aes(x=x,y=y)) + geom_point() 
```

## $ y = a + b \cdot x $

```r
model2 <- lm(y~x, simulated_data2)
summary(model2)
```

# Real-world example

```r
data <- read_csv('data/train.csv')
data %>% head
```

```r
lm <- lm(cnt ~ temp + factor(weathersit) + factor(weekday) + factor(season) + factor(yr), data)
summary(lm)
```

```r
new_row <- data[1,] 
new_row['temp'] <- 100
data2 <- rbind(data, new_row)
```

```r
lm <- lm(cnt ~ temp + factor(weathersit) + factor(weekday) + factor(season) + factor(yr), data2)
summary(lm)
```
